```python
# ============================================================================================
#                            Instagram Influencer Analysis - EDA
#=============================================================================================

#Import Libraries
import pandas as pd
import matplotlib.pyplot as plt

#Load Dataset
df =pd.read_csv("C:/Users/asus/Downloads/archive/top_insta_influencers_data.csv")
df.head()
df.info()

#Data Cleaning
df.isnull().sum()
df =df.dropna()
df =df.drop_duplicates()

#Data Preprocessing
df=df.rename(columns={'channel_info':'username'})
```

```python
#Top 5 Influencers Based on Followers
if df['followers'].dtype =='string':
    df['followers'] = df['followers'].str.upper()
    df['followers'] = df['followers'].str.replace('M','e6')
    df['followers'] = df['followers'].str.replace('K','e3')
    df['followers'] = df['followers'].astype(float)

top_followers = df.sort_values(by='followers', ascending=False).head(5)
print(top_followers[['username','followers']])
```

```python
# ============================================
# Top 5 Influencers Graph Based on Followers
# ============================================

if df['followers'].dtype =='string':
    df['followers'] = df['followers'].str.upper()
    df['followers'] = df['followers'].str.replace('M','e6')
    df['followers'] = df['followers'].str.replace('K','e3')
    df['followers'] = df['followers'].astype(float)

top_followers =df.sort_values(by='followers',ascending=False).head(5)

top_followers['followers_M'] = top_followers['followers']/1000000


top_followers.plot(x='username',y='followers_M',kind='bar')
plt.ylabel("followers(Millions)")
plt.title("Top 5 Influencers by Followers")
plt.xticks(rotation=0)
plt.show()
```

```python
# Observation:

# Top 5 influencers based on followers plotted successfully.
# Cristiano Ronaldo has the highest followers in the dataset.
```

```python
# =================================
# Followers Distribution Graph
# =================================

if df['followers'].dtype =='string':
    df['followers'] = df['followers'].str.upper()
    df['followers'] = df['followers'].str.replace('M','e6')
    df['followers'] = df['followers'].str.replace('K','e3')
    df['followers'] = df['followers'].astype(float)

plt.figure(figsize=(10,6))
followers_m = df['followers'] / 1000000

plt.hist(followers_m, bins=20, edgecolor='black')

plt.title("Followers Distribution")
plt.xlabel("Followers (in Millions)")
plt.ylabel("Number of Influencers")

plt.show()
```

```python
# Observation:

# Most of the influencers have followers below 100 Million.
# Very few influencers have extremely high followers.
# The distribution is right-skewed, meaning few accounts dominate with very large audience.
```

```python
# =============================================
# Followers vs Influence Score (Scatter Plot)
# =============================================

if df['followers'].dtype =='string':
    df['followers'] = df['followers'].str.upper()
    df['followers'] = df['followers'].str.replace('M','e6')
    df['followers'] = df['followers'].str.replace('K','e3')
    df['followers'] = df['followers'].astype(float)

followers_m = df['followers'] / 1000000

plt.figure(figsize=(10,6))
plt.scatter(followers_m, df['influence_score'])

plt.title("Followers vs Influence Score")
plt.xlabel("Followers (in Millions)")
plt.ylabel("Influence Score")

plt.show()
```

```python
# Observation:

# There is a general positive relationship between followers and influence score.
# Influencers with higher followers tend to have higher influence scores.
# However, some influencers show high followers but only moderate influence score, showing that followers alone do not fully determine influence.
```

```python
# ==========================================
# Country-wise Influencers Distribution 
# ==========================================


plt.figure(figsize=(10,6))
df['country'].value_counts().plot(kind='bar')

plt.title("Country-wise Influencers")
plt.xlabel("Country")
plt.ylabel("Number of Influencers")
plt.xticks(rotation=90)

plt.show()
```

```python
# Observation:

# United States has the highest number of influencers in the dataset.
# Other countries have significantly fewer influencers compared to the US.
# This shows that the influencer population is highly concentrated in a few countries.
```

```python
# Conclusion:

# The Exploratory Data Analysis (EDA) on the Instagram influencers dataset provided valuable insights.
# The analysis shows that followers and influence score are generally positively related, but not always directly proportional.
# The United States dominates the influencer population compared to other countries.
# Visualization techniques such as bar chart, histogram, and scatter plot helped in understanding data distribution, relationships, and patterns clearly.
# Overall, this analysis helps in understanding patterns in influencer popularity and engagement.
```
